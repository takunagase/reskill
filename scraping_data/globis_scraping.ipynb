for文を回す

```python
#d_listに空リストを作成する
d_list = []

# globisのURLを設定する
url_content = "https://hodai.globis.co.jp/learn-content/8467af48/"

# レスポンスを取得する
res = requests.get(url_content)

# 取得結果をsoupに格納する
soup = BeautifulSoup(res.text, 'html.parser', from_encoding='utf-8')

#カテゴリとして講義内容を取ってくる
contents = soup.find_all('article', class_='card')

# for文で各情報を取得する
for content in contents:
    #講義内容の詳細へとべるリンク
    a_tag = content.find('a', {'data-v-d28c2ef0': ''})
    url = a_tag.get('href')
    #使うときは以下の形でリンクに飛べる
    #print("https://hodai.globis.co.jp"+url)
    link = "https://hodai.globis.co.jp"+url
    #講義名
    text_area = content.find("div", class_="textArea")
    title = text_area.find("p", {"data-v-d28c2ef0": ""}).text.encode('utf-8').decode('utf-8').encode('ISO-8859-1').decode('utf-8')
    #難易度
    rank = text_area.find("span").text.encode('utf-8').decode('utf-8').encode('ISO-8859-1').decode('utf-8')
    #講義時間
    image_area = content.find("div", class_="imageArea")
    # もしimage_areaが見つからなかった場合は、Noneを返す
    if image_area:
        # image_area内からp要素を検索し、見つかればテキストを取得
        lecture_time_element = image_area.find("p")
        
        # もしlecture_time_elementが見つからなかった場合は、Noneを返す
        if lecture_time_element:
            lecture_time = lecture_time_element.text
        else:
            # p要素が見つからなかった場合の処理
            lecture_time = "Not available"
    else:
        # image_areaが見つからなかった場合の処理
        lecture_time = "Not available"
    
    #講義内容の詳細
    # linkにアクセスしてHTMLを取得
    detail = requests.get(link)

    # 取得したHTMLをBeautifulSoupで解析
    soup_detail = BeautifulSoup(detail.text, 'html.parser')

    #おすすめ
    recommend_information = soup_detail.find("div", class_="outline isHidden")
    recommends = recommend_information.find('span').text.encode('utf-8').decode('utf-8').encode('ISO-8859-1').decode('utf-8')
    
    #globisは講師情報が載っているものと載っていないものがある

    dic = {
        "link": link,
        "title": title,
        "rank": rank,
        "lecture_time": lecture_time,
        "recommends": recommends
    }

    d_list.append(dic)
```

```python
len(d_list)
#「もっと見る」の中の講義が取れていない
```

```python
max = range(0, len(contents))
df =pd.DataFrame(d_list)
```
